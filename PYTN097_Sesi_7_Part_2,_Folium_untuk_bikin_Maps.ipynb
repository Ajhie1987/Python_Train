# Folium
folium adalah package python yang akan kita gunakan untuk membuat visualisasi tipe peta

```python
# import package
import numpy as np
import pandas as pd

import folium
```

```python
folium.Map()
```

folium.Map() membuatkan peta dunia untuk kita yang interaktif

agar map nya lebih berguna, biasanya kita mau pilih fokus lokasi awal dimana. Contohnya, kita bisa minta Tokyo Tower sebagai lokasi awalnya.

Carilah di Google Maps lat dan long tempat ini.

Contohnya, Tokyo Tower ada di 35.65843227935634, 139.74535779450773

Berikutnya, kita set juga opsi untuk zoom value. karena zoom yang terlalu jauh jadi memperlihatkan seluruh dunia dan bukan lokasi kita.

```python
loc_tokyotower = (35.65843227935634, 139.74535779450773)

# simpan map nya ke variable supaya nanti bisa dipanggil lagi
map_tokyotower = folium.Map(location=loc_tokyotower, zoom_start=18)
map_tokyotower
```

Biasanya kita mau taruh marker untuk menandakan sesuatu di maps nya.

kita bisa ambil lokasi lat dan long dari sebuah dataset, kemudian kita loop per satu data, kita tambahkan marker ke map.

sebagai contoh, ada dataset tindakan kriminal di San Francisco. Kita bisa loop untuk setiap tindak kriminal, kita tambahkan marker warna merah dengan popup text detail tindakan kriminal apa yang terjadi disana.

## Contoh
di contoh kali ini, kita mau buat marker yang menandakan posisi setiap propinsi Jepang.

1. Kita import dataset
2. Kita buat dulu Map awal yang menunjukkan negara Jepang
3. untuk setiap baris data di dataset (1 propinsi) kita buatkan marker, dan kita tambahkan ke Map kita

```python
# import data
df_jp = pd.read_csv("https://raw.githubusercontent.com/stefkwan-h8/dataset/main/japan_prefecture_latlng.csv")
df_jp.head(3)
```

```python
df_jp.info()
```

```python
# buat map Japan
map_jp = folium.Map(location=loc_tokyotower, zoom_start=5)
map_jp
```

```python
# untuk setiap baris data di dataset kita === satu prefektur
# kita buatkan sebuah marker

map_jp = folium.Map(location=loc_tokyotower, zoom_start=5)

for row in df_jp[["prefecture_en", "lat", "long"]].values:
  [nama_pref, lat, long] = row

  mar = folium.Marker(location = (lat, long), tooltip = nama_pref)
  mar.add_to(map_jp)

map_jp
```

```python
# tiles mengganti stype map nya

map_jp = folium.Map(location=loc_tokyotower, zoom_start=5, tiles="Stamen Watercolor")

for row in df_jp[["prefecture_en", "lat", "long"]].values:
  [nama_pref, lat, long] = row

  mar = folium.Marker(location = (lat, long), tooltip = nama_pref)
  mar.add_to(map_jp)

map_jp
```
