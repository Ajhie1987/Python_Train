```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
import json
```

```python
iam = pd.read_excel('D:\File_Analisa_Python\IAM.xlsx')
iam

plt.style.use('ggplot')
```

```python
df_RO = iam[['POSISI','RO','JUMLAH']]

df_RO = df_RO.groupby(['POSISI','RO']).sum('JUMLAH').reset_index()

df_RO.pivot(index='POSISI', columns='RO', values='JUMLAH').plot(kind='area')
plt.title('Trend Data Anomali')
plt.xlabel('Posisi')
plt.tight_layout()
plt.show()
```

```python
df_anomali = iam[['PROSES_BISNIS', 'POSISI','JUMLAH']]
#df_anomali = df_anomali[df_anomali['POSISI'].dt.year == 2023 ]                           #Filter berdasarkan Tahun GLPAR1 (Posisi Saat Ini)

df_anomali = df_anomali.groupby(['POSISI', 'PROSES_BISNIS']).sum('JUMLAH').reset_index()

df_anomali = df_anomali.pivot(index='POSISI', columns='PROSES_BISNIS', values='JUMLAH')
df_anomali.plot(figsize=(12,5),
                marker='o')
plt.title('Trend Data Anomali berdasarkan Proses Bisnis')
plt.show()
```
