# Wordcloud
word cloud merangkum kumpulan text dengan cara visual, dimana text yang lebih sering muncul akan lebih besar ukurannya.

Kita akan gunakan package wordcloud buatan Amueller dari http://amueller.github.io/ untuk membuat word cloud

Contoh data yang akan kita gunakan adalah text novel "Alice in Wonderland" dari https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/alice_novel.txt

```python
# pip install wget
```

```python
# import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from wordcloud import WordCloud, STOPWORDS

import wget
import urllib.request
from PIL import Image
```

```python
# import data
url = 'https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/alice_novel.txt'
file_name = wget.download(url)
print(file_name)
```

```python
text_alice = open(file_name, 'r').read()
text_alice
```

\n adalah cara komputer menyimpan enter






```python
# bikin wordcloud
wc_alice = WordCloud()

wc_alice.generate(text_alice)
```

```python
# tampilkan wordcloud nya
plt.imshow(wc_alice)
```

hasil wordcloud yang kita buat:
- belum cantik, ada axis x dan y nya
- backgorund nya bisa kita ganti warna
- bisa kita perhalus dengan menggunakan opsi interpolation
- kurangi kata-kata tidak terpakai seperti "said"

```python
# lihat STOPWORDS yang diberikan
STOPWORDS
```

```python
print(type(STOPWORDS))
```

```python
# kita tambahkan "said" ke STOPWORDS dan kita ubah jadi tipe list
stopwords2 = list(STOPWORDS) + ["said"]
```

```python
# buat ulang wordcloud dengan opsi2 tambahan
# white background
# no axis
# tanpa "said" atau STOPWORDS lainnya

wc_alice2 = WordCloud(
    stopwords = stopwords2,
    background_color = "white"
)

wc_alice2.generate(text_alice)

# tambahkan juga opsi untuk menghaluskan hasil render gambar
# dan hapus axis
plt.imshow(wc_alice2, interpolation="bilinear")
plt.axis("off")
plt.show()
```

# wordcloud dengan mask option

mask adalah nama untuk bentuk gambar, contohnya gambar chat bubble ini https://raw.githubusercontent.com/stefkwan-h8/dataset/main/comment.png

kita akan buat wordcloud yang bentuknya menyesuaikan dengan bentuk gambar tersebut

```python
# import gambar untuk mask nya
url = 'https://raw.githubusercontent.com/stefkwan-h8/dataset/main/comment.png'
file_name = "bubble_mask.png"
urllib.request.urlretrieve(url, file_name)

bubble_mask = Image.open(file_name) # ini gambar beneran
# kita convert menjadi numpy array isi angka
bubble_mask_arr = np.array(bubble_mask)
bubble_mask_arr
```

```python
wc_alice3 = WordCloud(
    mask = bubble_mask_arr,
    background_color = 'white',
    stopwords = stopwords2
)

wc_alice3.generate(text_alice)

plt.imshow(wc_alice3, interpolation = "bilinear")
plt.axis("off")
plt.show()
```

# Regression
regression adalah cara untuk mencari pattern dari satu set data yang tersebar. Biasanya data bersifat angka continuous seperti jumlah uang, berat badan, tinggi badan, populasi, dll... angkanya tidak bertipe kategori seperti rating bintang 1-5, atau yes/no 0/1

Dari penyebaran angka continuous ini, kita mau mencari apakah ada pattern tertentu. Disanalah kita pakai teknik regression.

Paling mudah membayangkan regression dengan data bersifat 2 dimensi di sebuah scatter plot

```python
# import data dan rapiin dulu
df_gold = pd.read_csv("https://raw.githubusercontent.com/stefkwan-h8/dataset/main/gold.csv", parse_dates=['Date'])
df_gold.head(3)
```

```python
df_gold.info()
```

```python
df_gold['Tahun'] = df_gold['Date'].dt.year
df_gold.head(3)
```

```python
sns.regplot(
    x = df_gold['Tahun'],
    y = df_gold['Close'],
    data = df_gold,
    marker = "x"
    )
```
