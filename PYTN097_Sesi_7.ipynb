# Wordcloud
word cloud merangkum kumpulan text dengan cara visual, dimana text yang lebih sering muncul akan lebih besar ukurannya.

Kita akan gunakan package wordcloud buatan Amueller dari http://amueller.github.io/ untuk membuat word cloud

Contoh data yang akan kita gunakan adalah text novel "Alice in Wonderland" dari https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/alice_novel.txt

```python
# pip install wget
```

```python
# import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from wordcloud import WordCloud, STOPWORDS

import wget
```

```python
# import data
url = 'https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/alice_novel.txt'
file_name = wget.download(url)
print(file_name)
```

```python
text_alice = open(file_name, 'r').read()
text_alice
```

\n adalah cara komputer menyimpan enter






```python
# bikin wordcloud
wc_alice = WordCloud()

wc_alice.generate(text_alice)
```

```python
# tampilkan wordcloud nya
plt.imshow(wc_alice)
```

hasil wordcloud yang kita buat:
- belum cantik, ada axis x dan y nya
- backgorund nya bisa kita ganti warna
- bisa kita perhalus dengan menggunakan opsi interpolation
- kurangi kata-kata tidak terpakai seperti "said"

```python
# lihat STOPWORDS yang diberikan
STOPWORDS
```

```python
print(type(STOPWORDS))
```

```python
# kita tambahkan "said" ke STOPWORDS dan kita ubah jadi tipe list
STOPWORDS = list(STOPWORDS) + ["said"]
```

```python
# buat ulang wordcloud dengan opsi2 tambahan
# white background
# no axis
# tanpa "said" atau STOPWORDS lainnya

wc_alice2 = WordCloud(
    stopwords = STOPWORDS,
    background_color = "white"
)

wc_alice2.generate(text_alice)

# tambahkan juga opsi untuk menghaluskan hasil render gambar
# dan hapus axis
plt.imshow(wc_alice2, interpolation="bilinear")
plt.axis("off")
plt.show()
```
